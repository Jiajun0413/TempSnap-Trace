```python
import igraph as ig
import pandas as pd
from typing import List, Optional, Dict, Tuple


# --- create_date_mapping 函数保持不变 (使用上一版本从索引提取日期的逻辑) ---
def create_date_mapping(graphs: List[ig.Graph], chains: List[pd.DataFrame], debug: bool = False) -> Tuple[Dict[str, int], List[Tuple[int, Optional[pd.Timestamp], Optional[pd.Timestamp]]]]:
    """
    创建日期字符串到最合适图索引的映射。
    优先精确匹配图的最大日期，其次是小于等于目标日期的最近图，最后是大于目标日期的最早图。
    修改：从 chains 中 DataFrame 的索引提取日期。

    Args:
        graphs: igraph 图对象的列表。
        chains: DataFrame 列表，其索引包含日期信息。
        debug: 是否打印调试信息。

    Returns:
        一个元组，包含：
        - date_to_graph_idx: 日期字符串到图索引的字典。
        - graph_date_ranges: 每个图的索引及其最小/最大日期元组列表。
    """
    date_to_graph_idx = {}
    graph_date_ranges = []

    # 1. 获取每个图的日期范围 (基于图中节点的最大日期) - 这部分不变
    for idx, g in enumerate(graphs):
        min_date, max_date = None, None
        if g and g.vs and 'Date' in g.vs.attributes():
            graph_dates = []
            for v in g.vs:
                if 'Date' in v.attributes() and v['Date']:
                    try:
                        date = pd.to_datetime(str(v['Date']), errors='coerce')
                        if pd.notna(date):
                            graph_dates.append(date)
                    except Exception:
                         if debug:
                            print(f"警告: 图 {idx} 节点 {v.index} 日期解析失败: {v['Date']}")
            if graph_dates:
                try:
                    min_date = min(graph_dates)
                    max_date = max(graph_dates)
                except Exception as e:
                     if debug:
                         print(f"警告: 图 {idx} 计算日期范围失败: {e}")

        graph_date_ranges.append((idx, min_date, max_date))
        #if debug and min_date and max_date:
        #     print(f"  图索引[{idx}] -> 日期范围: {min_date.strftime('%Y-%m-%d')} 至 {max_date.strftime('%Y-%m-%d')}")
        #elif debug:
        #     print(f"  图索引[{idx}] -> 无有效日期")


    # 2. *** 修改: 从 DataFrame 索引收集需要映射的日期 ***
    all_dates_to_map = set()
    for i, chain_df in enumerate(chains): # 添加索引 i 用于调试
        if not chain_df.empty:
            # 尝试从索引解析日期
            original_index_values = chain_df.index.astype(str).tolist() # 获取原始索引值用于调试
            valid_dates = pd.to_datetime(chain_df.index, errors='coerce').dropna()
            parsed_count = len(valid_dates)
            added_count = 0
            newly_added_dates = set()

            if parsed_count > 0: # 只有在成功解析出日期时才继续
                for date_obj in valid_dates:
                    date_str = date_obj.strftime('%Y-%m-%d')
                    if date_str not in all_dates_to_map:
                        newly_added_dates.add(date_str)
                        all_dates_to_map.add(date_str)
                        added_count += 1


    # 3. 为每个收集到的日期找到最佳图索引 - 这部分不变
    valid_graph_ranges = [(idx, max_d) for idx, _, max_d in graph_date_ranges if max_d is not None]
    if not valid_graph_ranges:
        if debug:
            print("错误: 所有图中都没有有效的最大日期，无法创建映射。")
        if not all_dates_to_map:
             return {}, graph_date_ranges
        for date_str in sorted(all_dates_to_map):
            date_to_graph_idx[date_str] = 0
        return date_to_graph_idx, graph_date_ranges

    for date_str in sorted(all_dates_to_map):
        try:
            target_date = pd.to_datetime(date_str)
        except ValueError:
            if debug:
                print(f"警告: 无法解析日期字符串 '{date_str}'，跳过映射。")
            continue

        matching_idx = -1
        # ... (匹配逻辑不变) ...
        # 优先：精确匹配最大日期
        exact_matches = [idx for idx, max_d in valid_graph_ranges if max_d == target_date]
        if exact_matches:
            matching_idx = exact_matches[0] # 如果有多个精确匹配，取第一个
            #if debug: print(f"  日期 {date_str}: 精确匹配图索引 {matching_idx}")

        # 其次：小于等于目标日期的最近最大日期
        if matching_idx == -1:
            lesser_or_equal = [(idx, max_d) for idx, max_d in valid_graph_ranges if max_d <= target_date]
            if lesser_or_equal:
                matching_idx = max(lesser_or_equal, key=lambda item: item[1])[0]
                #if debug: print(f"  日期 {date_str}: 匹配小于等于的最近图索引 {matching_idx}")


        # 再次：大于目标日期的最早最大日期
        if matching_idx == -1:
            greater = [(idx, max_d) for idx, max_d in valid_graph_ranges if max_d > target_date]
            if greater:
                matching_idx = min(greater, key=lambda item: item[1])[0]
                #if debug: print(f"  日期 {date_str}: 匹配大于的最早图索引 {matching_idx}")

        # 最后：如果仍然没有找到，使用第一个有效图的索引
        if matching_idx == -1:
             if valid_graph_ranges: # 确保至少有一个有效图
                 matching_idx = valid_graph_ranges[0][0]
                 #if debug: print(f"  日期 {date_str}: 未找到合适匹配，使用第一个有效图索引 {matching_idx}")
             else:
                 # 如果没有有效图，无法分配索引，可以跳过或分配默认值（如-1）
                 if debug: print(f"  日期 {date_str}: 未找到合适匹配，且无有效图可供默认选择。")
                 continue # 跳过此日期的映射


        date_to_graph_idx[date_str] = matching_idx


    return date_to_graph_idx, graph_date_ranges


def get_community_nodes_df(
    community_id: int,
    date_str: str,
    graphs: List[ig.Graph],
    communities: List[List[List[str]]],
    final_chains: List[pd.DataFrame], # 添加 final_chains 用于日期映射
    debug: bool = False
) -> pd.DataFrame:
    """
    根据社区ID和日期查找社区内所有节点的属性信息，并返回DataFrame。
    使用完整的日期映射逻辑。
    修改：索引为 Mutations_str，列按指定顺序排列。

    Args:
        community_id: 目标社区的ID。
        date_str: 目标社区的日期字符串 (格式 'YYYY-MM-DD')。
        graphs: igraph 图对象的列表。
        communities: 社区列表，每个元素是一个时间点的社区划分 (节点名称列表的列表)。
        final_chains: 用于创建日期映射的 DataFrame 列表 (索引包含日期)。
        debug: 是否打印调试信息。

    Returns:
        一个 Pandas DataFrame，索引为 Mutations_str (或 node_name 作为后备)，
        列按 Date, ID, Clade, Lineage, Location, Ancestor, Ancestor_ID, ... 顺序排列。
        如果找不到社区或节点，则返回空的 DataFrame。
    """
    # 1. 创建日期到图索引的精确映射
    try:
        date_to_graph_idx, _ = create_date_mapping(graphs, final_chains, debug=debug)
        if not date_to_graph_idx:
             if debug:
                 print("错误: 未能成功创建日期映射。")
             return pd.DataFrame()
    except Exception as e:
        if debug:
            print(f"错误: 创建日期映射时发生异常: {e}")
        return pd.DataFrame()


    # 2. 获取目标日期的图索引
    if date_str not in date_to_graph_idx:
        if debug:
            # 查找最接近的可用日期进行提示
            available_dates = sorted(date_to_graph_idx.keys())
            closest_date = min(available_dates, key=lambda d: abs(pd.to_datetime(d) - pd.to_datetime(date_str))) if available_dates else "无"
            print(f"错误: 日期 '{date_str}' 未在创建的映射中找到。最接近的可用日期: {closest_date}。可用日期列表: {available_dates}")
        return pd.DataFrame()
    graph_idx = date_to_graph_idx[date_str]

    # 3. 检查索引和数据有效性
    if not (0 <= graph_idx < len(graphs) and graphs[graph_idx]):
        if debug:
            print(f"错误: 图索引 {graph_idx} 无效或图不存在 (日期 {date_str})。")
        return pd.DataFrame()
    graph = graphs[graph_idx]

    if not (0 <= graph_idx < len(communities) and communities[graph_idx]):
        if debug:
            print(f"错误: 社区列表索引 {graph_idx} 无效 (日期 {date_str})。")
        return pd.DataFrame()
    community_list_for_date = communities[graph_idx]

    if not (0 <= community_id < len(community_list_for_date)):
        if debug:
            print(f"错误: 社区ID {community_id} 在日期 {date_str} (图索引 {graph_idx}) 的社区列表中越界。列表长度: {len(community_list_for_date)}")
        return pd.DataFrame()
    community_node_names = community_list_for_date[community_id]

    if not community_node_names:
         if debug:
             print(f"警告: 社区 {community_id} 在日期 {date_str} (图索引 {graph_idx}) 为空。")
         return pd.DataFrame()

    # 4. 提取节点属性
    all_attributes_keys = set() # 用于收集所有出现过的属性名
    temp_node_data = {} # 临时存储 {index_value: {attr: value}}
    index_values = []   # 存储最终的索引值

    for node_name in community_node_names:
        try:
            graph_node = graph.vs.find(name=node_name)
            node_attrs = graph_node.attributes()

            # *** 修改: 使用 Mutations_str 作为索引 ***
            index_key = node_attrs.get('name')
            if index_key is None:
                index_key = node_name # 使用原始 node_name 作为后备索引
                if debug:
                    print(f"警告: 节点 '{node_name}' 缺少 'Mutations_str' 属性，将使用节点名称作为索引。")

            # 避免添加重复的 'node_name' 列 (如果需要原始名称，可以保留在 node_attrs 中)
            # node_attrs['original_node_name'] = node_name # 如果需要保留原始名称

            # 检查索引是否已存在 (Mutations_str 可能不唯一)
            if index_key in temp_node_data:
                 if debug:
                     print(f"警告: 索引值 '{index_key}' (来自节点 '{node_name}') 已存在，此节点将被覆盖。请检查 'Mutations_str' 的唯一性。")
                 # 可以选择如何处理重复：跳过、合并、添加后缀等。目前是覆盖。

            temp_node_data[index_key] = node_attrs
            index_values.append(index_key) # 收集索引值
            all_attributes_keys.update(node_attrs.keys())

        except ig.VertexNotFound:
            if debug:
                print(f"警告: 在图 {graph_idx} (日期 {date_str}) 中未找到节点 '{node_name}' (社区 {community_id})。")
        except Exception as e:
            if debug:
                print(f"错误: 提取节点 '{node_name}' 属性时出错: {e}")

    # 5. 构建 DataFrame
    if not temp_node_data:
        if debug:
            print(f"未能在图 {graph_idx} 中找到社区 {community_id} (日期 {date_str}) 的任何有效节点。")
        return pd.DataFrame()

    # 准备数据和列
    final_data = []
    # 使用收集到的 index_values 保持顺序，并处理可能的重复（如果需要去重，可以在此处理）
    unique_index_values = pd.unique(index_values).tolist() # 获取唯一索引并保持顺序

    # 移除内部使用的 'node_name' (如果之前添加了) 或其他不需要的列
    attributes_to_include = sorted([k for k in all_attributes_keys if k != 'node_name' and k != 'name']) # igraph 内部可能有 'name'

    for idx_val in unique_index_values:
        row_data = {}
        node_attrs = temp_node_data[idx_val]
        for key in attributes_to_include:
            row_data[key] = node_attrs.get(key, None) # 使用 get 处理缺失属性
        final_data.append(row_data)

    df = pd.DataFrame(final_data, index=unique_index_values)
    df.index.name = 'Mutations_str' # 设置索引名称

    # *** 修改: 调整列顺序 ***
    desired_order = ['Date', 'ID', 'Clade', 'Lineage', 'Location', 'Ancestor', 'Ancestor_ID']
    # 找出实际存在于 DataFrame 中的列，并按期望顺序排列
    ordered_cols = [col for col in desired_order if col in df.columns]
    # 找出其他剩余的列
    other_cols = [col for col in df.columns if col not in ordered_cols]
    # 合并列顺序
    final_col_order = ordered_cols + sorted(other_cols) # 对其他列进行排序

    df = df[final_col_order]

    return df

# 调用新函数
nodes_df = get_community_nodes_df(
    community_id=12,
    date_str='2020-06-25',
    graphs=graphs,
    communities=communities,
    final_chains=tracking_chains_africa, # 传入 final_chains
    debug=True
)
```

```python
def check_graph_weights(g, verbose=True):
    """
    检查图对象中是否存在点权重和边权重，并提供统计信息
    
    参数:
    - g: igraph图对象
    - verbose: 是否打印详细信息
    
    返回:
    - (has_vertex_weight, has_edge_weight): 表示是否有点权重和边权重的元组
    """
    import numpy as np
    
    # 检查点权重
    has_vertex_weight = 'wdks' in g.vs.attributes()
    
    # 检查边权重
    has_edge_weight = 'weight' in g.es.attributes()
    
    if verbose:
        print(f"图对象信息: {g.vcount()}个节点, {g.ecount()}条边")
        print(f"节点是否有权重: {has_vertex_weight}")
        print(f"边是否有权重: {has_edge_weight}")
        
        # 如果有点权重，提供统计信息
        if has_vertex_weight:
            weights = np.array(g.vs['wdks'])
            print(f"节点权重统计: 最小={weights.min():.4f}, 最大={weights.max():.4f}, "
                  f"平均={weights.mean():.4f}, 中位数={np.median(weights):.4f}")
            print(f"节点权重分布: [0.0-0.2]:{np.sum(weights <= 0.2)}, "
                  f"(0.2-0.3]:{np.sum((weights > 0.2) & (weights <= 0.3))}, "
                  f"(0.3-0.4]:{np.sum((weights > 0.3) & (weights <= 0.4))}, "
                  f"(0.4-0.5]:{np.sum((weights > 0.4)& (weights <= 0.5))}, ")
        
        # 如果有边权重，提供统计信息
        if has_edge_weight:
            weights = np.array(g.es['evo_weight'])
            print(f"边权重统计: 最小={weights.min():.4f}, 最大={weights.max():.4f}, "
                  f"平均={weights.mean():.4f}, 中位数={np.median(weights):.4f}")
            print(f"边权重分布: [0.0-0.2]:{np.sum(weights <= 0.2)}, "
                  f"(0.2-0.5]:{np.sum((weights > 0.2) & (weights <= 0.5))}, "
                  f"(0.5-0.8]:{np.sum((weights > 0.5) & (weights <= 0.8))}, "
                  f"(0.8-1.0]:{np.sum(weights > 0.8)}")
    
    return has_vertex_weight, has_edge_weight


for i, g in enumerate(backbone_networks):
    if g is not None and g.vcount() > 0:
        print(f"检查图 {i}:")
        check_graph_weights(g)
        print("-------------------")

```
